```python
#We are going to show the total number of batteries, and Is the battery removable or not, and showing the maximum and the minimum number of battery capacity in MHz.
import re
# converting the remove able battery to the   'True' and 'False'
def removable(battery):
    if battery[0] == 'Removable':
        return True
    elif battery[0] == 'Non-removable':
        return False
    else:
        return np.nan

a = df_top.battery.astype(str).apply(lambda x:(re.findall('(Non-removable|Removable|)\s*(Li-Ion|Li-Po|Li-ion|Li-po|NiMH|)\s*(\d*)',x)))

df_top['battery_removable'] = a.apply(lambda x: removable(x[0]))
df_top['battery_type'] = a.apply(lambda x: (x[0])[1])
df_top['battery_mah'] = a.apply(lambda x: ((x[0])[2]))
df_top['battery_mah']= pd.to_numeric(df_top['battery_mah'],errors='coerce')
#df_top.drop(columns=['battery'],inplace=True)
print('the minimum battery capacity : ',df_top['battery_mah'].min(),' mAh')
print('the minimum battery capacity in : ',df_top['battery_mah'].max(),' mAh')
counts =df_top['battery_removable'].value_counts()

explode = [0.3,0.4]
textprops = {"fontsize":12}
display(counts)
plt.pie(counts,labels=counts.keys(),radius=3, explode = explode,autopct = "%0.2f%%",shadow = True,startangle = 270,textprops =textprops)
plt.legend()
plt.show()

```
